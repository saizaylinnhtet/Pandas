### Data Cleaning

```python
# Load Data Jobs Dataset
dataset = load_dataset('lukebarousse/data_jobs')
dataset
```

```python
df = dataset['train'].to_pandas()
```

```python
df.info()
```

```python
df.job_posted_date
```

```python
df.job_posted_date[0]
```

```python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
df.info()
```

```python
df['job_posted_month'].head(10)
```

```python
df.sort_values('job_posted_date')
```

```python
df.sort_values('job_posted_date', inplace=True)
df
```

```python
df.drop(labels='salary_hour_avg', axis=1, inplace=True)
df
```

```python
df.dropna(subset=['salary_year_avg'], inplace=True)
```
